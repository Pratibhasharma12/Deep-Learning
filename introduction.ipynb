# Perceptron
## What is a Perceptron?

A Perceptron is the simplest form of a neural network.
It represents a single neuron.
- It takes inputs, applies weights, adds bias, and produces output.

Example:
Think of a perceptron like a decision maker.
You want to decide if a student will pass or fail.
Inputs:
- Hours studied
- Attendance
Perceptron:
- Assigns importance (weights)
- Combines inputs
- Gives output (pass/fail)
It learns a linear decision boundary

Formula: 
- y=w1​x1​+w2​x2​+⋯+wn​xn​+b

Where:
- x = inputs
- w = weights
- b = bias
- y = output

This is basically a weighted sum
Then we apply an activation function:

- y=activation(wx+b)
## Geometric View

Perceptron creates a line (2D) or plane (3D)

It separates data into two classes

Example:
- Above line → Class 1
- Below line → Class 0

This is called a linear classifier
## Limitation of Perceptron

Can only solve linearly separable problems

Example:
- Cannot solve XOR problem
This led to Multi-Layer Neural Networks

```python
import numpy as np

class Perceptron:
    
    def __init__(self, lr=0.1, epochs=10):
        self.lr = lr
        self.epochs = epochs
    
    def fit(self, X, y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for _ in range(self.epochs):
            for i in range(len(X)):
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_pred = 1 if linear_output >= 0 else 0
                
                # Update rule
                self.weights += self.lr * (y[i] - y_pred) * X[i]
                self.bias += self.lr * (y[i] - y_pred)

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return [1 if i >= 0 else 0 for i in linear_output]


# Example dataset
X = np.array([[2,3],[1,1],[2,1],[3,2]])
y = np.array([1,0,0,1])

model = Perceptron()
model.fit(X,y)

print("Predictions:", model.predict(X))
```

Code Explanation
- Initializes weights and bias
- Loops over dataset multiple times (epochs)
- Calculates prediction
- Updates weights using error

Perceptron Learning Rule
- Weight Update Formula:

w = w + lr * (y_true - y_pred) * x

If prediction is wrong → weights adjust
If correct → no change

## What are Activation Functions?

Activation functions are mathematical functions applied to the output of a neuron.

👉 They decide whether a neuron should be activated or not.

Without activation function → neural network behaves like a linear model (useless for complex problems)

Why Activation Functions are Needed?

Imagine stacking multiple perceptrons (layers).

If no activation function is used:

Entire network = just a linear equation

y = w1x1 + w2x2 + ...
No matter how many layers you add → still not linear 

### Key Insight:
Activation functions introduce NON-LINEARITY

This allows the network to learn complex patterns like:
- Images
- Speech
- Text

## TYPES OF ACTIVATION FUNCTIONS
1. Sigmoid Function

Range:
0 to 1

Use:
Binary classification

Advantages:
- Smooth output
- Interpretable as probability

Disadvantages:
- Vanishing gradient problem
- Slow learning
- Not zero-centered

Formula:

σ(x)=1/1+e^−x

2. Tanh Function

Range:
-1 to 1

Advantages:
- Zero-centered
- Better than sigmoid

Disadvantages:
- Still suffers from vanishing gradient

Formula:

tanh(x)=e^x+e^−x/e^x−e^−x

3. ReLU (Rectified Linear Unit)

Formula:
f(x) = max(0, x)

Advantages:
- Fast computation
- No vanishing gradient (for positive values)
- Most widely used

Disadvantages:
- Dead neuron problem (output = 0 for negative values)

## Comparison of Activation Functions

| Function | Range | Problem |
|---------|------|--------|
| Sigmoid | (0,1) | Vanishing gradient |
| Tanh | (-1,1) | Vanishing gradient |
| ReLU | (0,∞) | Dead neurons |

 Best Practice:
- Use ReLU in hidden layers
- Use Sigmoid in output (binary classification)

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

x = np.array([-2, -1, 0, 1, 2])

print("Sigmoid:", sigmoid(x))
print("Tanh:", tanh(x))
print("ReLU:", relu(x))
```

## What is Forward Propagation?

Forward propagation is the process of passing input data through the neural network to get the output.

It is the "prediction step" of a neural network.

Example:

Think of a neural network like a pipeline:

Input → Layer 1 → Layer 2 → Output

At each layer:
1. Multiply inputs with weights
2. Add bias
3. Apply activation function

👉 Output of one layer becomes input to next layer

This continues until final output is produced.

Steps in Forward Propagation

1. Take input data (X)
2. Multiply with weights (W)
3. Add bias (b)
4. Apply activation function
5. Pass output to next layer
6. Repeat until final output

This is done for every input sample

Key Insight

Each layer transforms the data into a better representation.

Example:
- Layer 1 → edges
- Layer 2 → shapes
- Layer 3 → objects

👉 This is hierarchical learning

```python
import numpy as np

# Input (1 sample, 2 features)
X = np.array([[1, 2]])

# Weights
W1 = np.random.randn(2, 3)
W2 = np.random.randn(3, 1)

# Bias
b1 = np.zeros((1, 3))
b2 = np.zeros((1, 1))

# Activation function
def relu(x):
    return np.maximum(0, x)

# Forward propagation
Z1 = np.dot(X, W1) + b1
A1 = relu(Z1)

Z2 = np.dot(A1, W2) + b2
A2 = Z2  # output layer (no activation for simplicity)

print("Final Output:", A2)
```

```python
import numpy as np

class SimpleNN:
    
    def __init__(self):
        self.W1 = np.random.randn(2, 3)
        self.W2 = np.random.randn(3, 1)
        self.b1 = np.zeros((1, 3))
        self.b2 = np.zeros((1, 1))
    
    def relu(self, x):
        return np.maximum(0, x)
    
    def forward(self, X):
        self.Z1 = np.dot(X, self.W1) + self.b1
        self.A1 = self.relu(self.Z1)
        
        self.Z2 = np.dot(self.A1, self.W2) + self.b2
        return self.Z2


# Test
model = SimpleNN()
X = np.array([[1, 2]])

print("Output:", model.forward(X))
```

## What is Backpropagation?

Backpropagation is the process of updating weights in a neural network to reduce error.
It is the "learning step" of a neural network.

Forward Propagation → Prediction  
Backpropagation → Learning

Example:

Think like this:

1. Model makes a prediction
2. Compare with actual answer → calculate error
3. Adjust weights to reduce error

Just like learning from mistakes

Example:
If prediction is wrong → adjust weights  
If correct → keep weights similar
## 🔁 Training Flow

1. Forward Propagation → get prediction
2. Compute Loss (error)
3. Backpropagation → compute gradients
4. Update weights

Repeat many times (epochs)

## Loss Function
Formula:

L=n1​∑(ytrue​−ypred​)2

Loss measures how wrong the model is.

Goal:
Minimize loss
## What is Gradient?

Gradient = rate of change of loss w.r.t weights

Tells:
- which direction to move
- how much to change weights
We use derivatives

## GRADIENT DESCENT UPDATE RULE

Formula:

w=w−η∂L/∂w

Where:
- η (eta) = learning rate
- ∂L/∂w = gradient

👉 Move in opposite direction of gradient
## 🔥 Chain Rule

Backpropagation uses chain rule from calculus.

👉 It breaks complex derivative into smaller parts

Example:
dL/dW = dL/dA × dA/dZ × dZ/dW

👉 This is why it's called "back-propagation"

```python
import numpy as np

# Input
X = np.array([[1, 2]])
y = np.array([[1]])

# Initialize weights
W1 = np.random.randn(2, 2)
W2 = np.random.randn(2, 1)

lr = 0.1

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

# Training loop
for epoch in range(100):

    # Forward
    Z1 = np.dot(X, W1)
    A1 = relu(Z1)
    
    Z2 = np.dot(A1, W2)
    y_pred = Z2

    # Loss
    loss = np.mean((y - y_pred) ** 2)

    # Backward
    dZ2 = (y_pred - y)
    dW2 = np.dot(A1.T, dZ2)

    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = np.dot(X.T, dZ1)

    # Update
    W2 -= lr * dW2
    W1 -= lr * dW1

print("Final Prediction:", y_pred)
```

```python
# Momentum
w = 0
v = 0
lr = 0.1
beta = 0.9

for i in range(20):
    grad = 2 * (w - 3)
    v = beta * v + (1 - beta) * grad
    w = w - lr * v

print("Optimized w:", w)
```

```python
# ADAM
w = 0
m, v = 0, 0
lr = 0.1
beta1, beta2 = 0.9, 0.999
epsilon = 1e-8

for i in range(50):
    grad = 2 * (w - 3)
    
    m = beta1 * m + (1 - beta1) * grad
    v = beta2 * v + (1 - beta2) * (grad**2)
    
    w = w - lr * m / (np.sqrt(v) + epsilon)

print("Optimized w:", w)
```

## What is Weight Initialization?

Weight Initialization = assigning initial values to weights before training.

Important because:
It affects how fast and how well the model learns

Intuition

Think of training like starting a journey:

- Good initialization → reach destination faster
- Bad initialization → get stuck or move slowly

Initial weights decide learning behavior

## Problems with Bad Initialization

### 1. All weights = 0
- All neurons learn same thing 
- No learning happens

### 2. Very large weights
- Activations become huge
- Leads to exploding gradients 

### 3. Very small weights
- Signals shrink
- Leads to vanishing gradients 

## Vanishing Gradient Problem

During backpropagation:
- Gradients become very small
- Weight updates become negligible

Model stops learning

Example:

Imagine passing a signal through many layers:

Each layer reduces the signal slightly

After many layers:
Signal becomes almost zero 

So weights don't update

Happens in:

- Deep networks
- Sigmoid / Tanh activation

Because derivatives are small

## Exploding Gradient Problem

Gradients become very large during backpropagation.

 Causes:
- Huge weight updates
- Unstable training

1. ReLU Activation
### Use ReLU instead of Sigmoid/Tanh

- Prevents vanishing gradient (for positive values)
- Faster training

 Most commonly used activation
 2. Proper Weight Initialization
 Xavier Initialization (for tanh/sigmoid)

- Keeps variance stable
- Prevents vanishing/exploding
 He Initialization (for ReLU)

- Best for ReLU networks
- Maintains signal flow
 3. Batch Normalization
### Batch Normalization

- Normalizes inputs of each layer
- Keeps values in stable range

 Benefits:
- Faster training
- Reduces gradient problems
 4. Gradient Clipping
### Gradient Clipping

- Limits gradient size

 Used to:
- Prevent exploding gradients

#  Regularization

Regularization helps prevent overfitting.

 Idea:
Add penalty to large weights

 Goal:
Make model simpler and generalize better

##  L2 Regularization (Ridge)
Formula:

L=Loss+λ∑w2

- Penalizes squared weights

###  Effect:
- Reduces weight values
- Prevents overfitting

 Keeps all features but shrinks them

 ## L1 Regularization (Lasso)
Formula:

L=Loss+λ∑∣w∣

- Penalizes absolute weights

###  Effect:
- Some weights become zero

 Performs feature selection

## ⚖️ L1 vs L2

| Feature | L1 | L2 |
|--------|----|----|
| Effect | Sparse weights | Small weights |
| Feature Selection | Yes | No |
| Use Case | High-dimensional data | General use |

```python
# L2 REGULARIZATION
import numpy as np

def l2_loss(y_true, y_pred, weights, lambda_):
    mse = np.mean((y_true - y_pred) ** 2)
    l2_penalty = lambda_ * np.sum(weights**2)
    return mse + l2_penalty

print("L2 Loss:", l2_loss([1,2], [1.5,1.8], np.array([0.5, -0.3]), 0.1))
```

```python
# DROPOUT (SIMULATION)
import numpy as np

def dropout(x, rate=0.5):
    mask = np.random.binomial(1, 1-rate, size=x.shape)
    return x * mask

x = np.array([1,2,3,4,5])

print("After Dropout:", dropout(x))
```

## Artificial Neural Networks (ANN)

An Artificial Neural Network (ANN) is a computational model inspired by the human brain. It consists of layers of interconnected neurons that process input data and produce output through weighted transformations and activation functions.

Structure of ANN:
- Input Layer: Takes input features
- Hidden Layers: Perform transformations and feature learning
- Output Layer: Produces final prediction

Each neuron performs:
Z = XW + b
A = activation(Z)

Where:
- X = input
- W = weights
- b = bias
- Z = linear combination
- A = activated output

Flow of ANN:
Input → Hidden Layer(s) → Output

Each layer learns increasingly complex representations of data.

Key concept:
ANN is capable of learning non-linear relationships using activation functions like ReLU, sigmoid, etc.

Important properties:
- Learns from data using backpropagation
- Improves performance using gradient descent
- Requires proper initialization and scaling

Limitations:
- Requires large data
- Computationally expensive
- Hard to interpret (black-box)

## Forward and Backward Propagation in ANN

Forward Propagation:
- Input is passed layer by layer
- Each layer computes Z and A
- Final output is prediction

Mathematically:
Z1 = XW1 + b1
A1 = f(Z1)

Z2 = A1W2 + b2
A2 = f(Z2)

Backward Propagation:
- Compute loss between predicted and actual output
- Calculate gradients using chain rule
- Update weights to minimize loss

Weight update:
W = W - learning_rate * gradient

Training Loop:
1. Initialize weights
2. Perform forward propagation
3. Compute loss
4. Perform backpropagation
5. Update weights
6. Repeat for multiple epochs

## Key Design Choices in ANN

1. Number of layers:
- More layers → more complex learning
- Too many → overfitting

2. Number of neurons:
- Controls model capacity

3. Activation functions:
- ReLU for hidden layers
- Sigmoid/Softmax for output

4. Loss function:
- MSE for regression
- Cross-entropy for classification

5. Optimizer:
- Adam (most commonly used)

6. Learning rate:
- Controls speed of learning

7. Regularization:
- Dropout, L2 to prevent overfitting

```python
import numpy as np

class ANN:
    def __init__(self):
        self.W1 = np.random.randn(2, 3)
        self.b1 = np.zeros((1, 3))
        self.W2 = np.random.randn(3, 1)
        self.b2 = np.zeros((1, 1))
        self.lr = 0.1

    def relu(self, x):
        return np.maximum(0, x)

    def relu_derivative(self, x):
        return (x > 0).astype(float)

    def forward(self, X):
        self.Z1 = np.dot(X, self.W1) + self.b1
        self.A1 = self.relu(self.Z1)
        self.Z2 = np.dot(self.A1, self.W2) + self.b2
        return self.Z2

    def backward(self, X, y, output):
        dZ2 = (output - y)
        dW2 = np.dot(self.A1.T, dZ2)
        db2 = np.sum(dZ2, axis=0, keepdims=True)

        dA1 = np.dot(dZ2, self.W2.T)
        dZ1 = dA1 * self.relu_derivative(self.Z1)
        dW1 = np.dot(X.T, dZ1)
        db1 = np.sum(dZ1, axis=0, keepdims=True)

        self.W2 -= self.lr * dW2
        self.b2 -= self.lr * db2
        self.W1 -= self.lr * dW1
        self.b1 -= self.lr * db1

# Training
X = np.array([[1,2],[2,1],[3,3]])
y = np.array([[1],[0],[1]])

model = ANN()

for epoch in range(500):
    output = model.forward(X)
    model.backward(X, y, output)

print("Final Output:\n", model.forward(X))
```

## Understanding the Code

- We initialize weights randomly
- Forward pass computes predictions
- Backward pass computes gradients
- We update weights using gradient descent

Key Observations:
- Model improves with more epochs
- Learning depends on correct gradient flow
- Activation functions introduce non-linearity

Common Issues:
- Vanishing gradients (use ReLU)
- Overfitting (use regularization)
- Poor convergence (tune learning rate)

This implementation represents a basic fully connected neural network and forms the foundation for advanced architectures like CNNs and RNNs.

## Convolutional Neural Networks (CNN)

Convolutional Neural Networks are specialized neural networks designed for processing grid-like data such as images.

Why CNN instead of ANN for images:
- Images have spatial structure (height, width, channels)
- ANN flattens images → loses spatial information
- CNN preserves spatial relationships

Key idea:
Instead of fully connecting every neuron, CNN uses filters (kernels) to scan the image and extract features.

Input (Image) → Convolution → Activation → Pooling → Fully Connected → Output

CNN automatically learns:
- Edges
- Textures
- Shapes
- Objects

This hierarchical feature learning makes CNN very powerful for image-related tasks.

## Convolution Operation

Convolution is the core operation of CNN.

A small matrix (filter/kernel) slides over the image and performs element-wise multiplication and summation.

Mathematically:
Output[i,j] = sum(Filter * Image patch)

Important terms:

1. Filter (Kernel):
- Small matrix (e.g., 3x3)
- Detects specific features like edges

2. Stride:
- Number of steps filter moves
- Larger stride → smaller output

3. Padding:
- Adding zeros around image
- Prevents shrinking of image size

Types:
- Valid padding (no padding)
- Same padding (output size same as input)

Output size formula:
((N - F + 2P) / S) + 1

Where:
- N = input size
- F = filter size
- P = padding
- S = stride

## Activation and Pooling

After convolution, activation function is applied:
- ReLU is most common

Purpose:
- Introduce non-linearity

Pooling Layer:
Reduces spatial dimensions

Types:

1. Max Pooling:
- Takes maximum value
- Preserves important features

2. Average Pooling:
- Takes average value

Benefits of pooling:
- Reduces computation
- Prevents overfitting
- Makes model invariant to small changes

## CNN Architecture Flow

1. Input Image (e.g., 28x28x1)

2. Convolution Layer
- Apply multiple filters
- Output = feature maps

3. Activation (ReLU)

4. Pooling Layer
- Reduce size

5. Repeat Conv + Pool multiple times

6. Flatten Layer
- Convert 2D → 1D

7. Fully Connected Layer
- Perform classification

8. Output Layer
- Softmax (multi-class)
- Sigmoid (binary)

Key Insight:
Earlier layers → simple features (edges)
Deeper layers → complex features (objects)

## Why CNN Works Better Than ANN for Images

ANN:
- Treats each pixel independently
- Large number of parameters
- Loses spatial relationships

CNN:
- Shares weights (same filter across image)
- Detects patterns locally
- Fewer parameters
- Better generalization

Example:
Image 64x64 → ANN would need 4096 inputs
CNN reduces parameters significantly

```python
import numpy as np

# Simple image (5x5)
image = np.array([
    [1,2,3,0,1],
    [0,1,2,3,1],
    [1,2,1,0,0],
    [0,1,3,2,1],
    [1,0,2,1,0]
])

# Filter (3x3)
kernel = np.array([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1]
])

output = []

for i in range(image.shape[0] - 2):
    row = []
    for j in range(image.shape[1] - 2):
        region = image[i:i+3, j:j+3]
        value = np.sum(region * kernel)
        row.append(value)
    output.append(row)

print(np.array(output))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

## Recurrent Neural Networks (RNN)

Recurrent Neural Networks are designed for sequential data where order matters.

Examples of sequential data:
- Text (sentences)
- Time series (stock prices, weather)
- Speech signals

Problem with ANN/CNN:
- They assume inputs are independent
- Cannot remember previous information

RNN solves this by introducing "memory"

Core idea:
Output at time t depends on:
- Current input (Xt)
- Previous hidden state (Ht-1)

This allows RNN to capture temporal dependencies.

Mathematical representation:
Ht = f(Wx * Xt + Wh * Ht-1 + b)
Yt = Wy * Ht

Where:
- Xt = input at time t
- Ht = hidden state (memory)
- Yt = output

## Working of RNN (Unrolled View)

RNN can be visualized as repeating the same network over time.

Sequence example:
Input: "I love AI"

Step-by-step:
t1 → "I"
t2 → "love"
t3 → "AI"

At each step:
- Takes current word
- Combines with past memory
- Produces output

Key property:
Same weights are reused at every time step

This is called parameter sharing

## Types of RNN Architectures

1. One-to-One
- Simple neural network

2. One-to-Many
- Image → caption generation

3. Many-to-One
- Sentiment analysis

4. Many-to-Many
- Language translation
- Chatbots

These architectures depend on input-output sequence requirements

## LSTM (Long Short-Term Memory)

LSTM is an improved version of RNN designed to handle long-term dependencies.

It introduces gates to control information flow.

Key idea:
Decide what to:
- Keep
- Forget
- Output

## LSTM Gates

1. Forget Gate
- Decides what information to discard

2. Input Gate
- Decides what new information to store

3. Output Gate
- Decides what to output

These gates use sigmoid activation:
- Output between 0 and 1
- Controls flow of information

Cell state:
Acts as long-term memory

```python
import numpy as np

# Example sequence
X = [1, 2, 3]

# Initialize weights
Wx = 0.5
Wh = 0.8
b = 0.1

h = 0  # initial hidden state

for x in X:
    h = np.tanh(Wx * x + Wh * h + b)
    print("Hidden state:", h)
```
